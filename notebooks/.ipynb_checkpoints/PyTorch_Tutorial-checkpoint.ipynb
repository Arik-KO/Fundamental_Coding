```python
numpy_matrix = np.array([
    [1,4,5], [2,7,9]
])
print(numpy_matrix.shape)
print(numpy_matrix.dtype)
tensor_matrix = torch.from_numpy(numpy_matrix)
print(tensor_matrix.shape)
print(tensor_matrix.dtype)
```
