```python
from torch.utils.data import Dataset, DataLoader
```

```python
class SquareDataset(Dataset):
    def __init__(self, length=10):
        self.data = torch.arange(length)

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        x = self.data[idx]
        y = x ** 2
        return x,y
        
```

```python
experiment_dataset = SquareDataset(length=10)
dataloader_experimental = DataLoader(experiment_dataset, batch_size = 4, shuffle = True)
```

```python
for batch_idx, (x,y) in enumerate(dataloader_experimental):
    print(f"batch : {batch_idx+1}")
    print(f"x : {x}")
    print(f"y : {y}")
```
