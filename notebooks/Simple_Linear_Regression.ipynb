```python
import torch
import torch.nn as nn
```

```python
torch.manual_seed(42)
X = torch.linspace(0,10,100).unsqueeze(1)
print(f"shape of input X: {X.shape}")
```

```python
true_w, true_b = 2.0, 1.0
```

```python
y = true_w * X + true_b + 0.3*torch.rand_like(X)
```

```python
y.shape
```

```python
linear_layer = nn.Linear(in_features=1, out_features=1)
```

```python
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(linear_layer.parameters(), lr = 1e-3)
```

```python
num_epochs = 200
```

```python
for epoch in range(num_epochs):
    linear_layer.train()
    y_pred = linear_layer(X)
    loss = criterion(y_pred,y)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    if (epoch + 1) % 10 == 0:
        print(f"Epoch {epoch+1:3d}, loss = {loss.item():.4f}")
```

```python
linear_layer.eval()

with torch.no_grad():
    y_pred_train = linear_layer(X)
    final_train_loss = criterion(y_pred_train, y)
    print("Final train loss:", final_train_loss.item())
```
