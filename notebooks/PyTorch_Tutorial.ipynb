# torch.gather()

```python
# using gather function we can fetch particular elements from each row(dim=1) or column(dim=0)
```

```python
var_a = torch.rand((3,4))
var_a
```

```python
indecies_to_fetch = torch.tensor([
    [2],[1],[3]
])
```

```python
torch.gather(var_a,dim=1,index= indecies_to_fetch)
```

```python
col_indecies = torch.tensor([
    [0],[1],[2],[2]
])
```

```python
torch.gather(var_a, dim =0, index = col_indecies)
```
